## Creation of input and output features (Task C)

```python
features = df.drop('Attrition', axis=1)
features
```

```python
labels = df['Attrition']
labels
```

## Conversion of features into numeric values (Task D)

```python
non_numeric_columns = features.select_dtypes(exclude=['number']).columns
non_numeric_columns
```

```python
numeric_collums = features.select_dtypes(include=['number']).columns
numeric_collums
```

```python
for col in non_numeric_columns:
    features[col] = features[col].astype('category')

features.dtypes
```

```python
features = pd.get_dummies(features)
features
```

```python
features.dtypes
```

```python
if features.select_dtypes(include=['bool']).any().any():
    features = features.astype('int64')
features.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
labels = encoder.fit_transform(labels)
labels = pd.DataFrame(labels, columns=['Attrition'])
labels
```

## Scaling of the features (Task E)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

def scaling(scalingmethod):
    if scalingmethod == 'minmax':
        return MinMaxScaler()
    elif scalingmethod == 'standard':
        return StandardScaler()
    else:
        raise ValueError("Invalid scaling method. Choose 'minmax' or 'standard'.")
```

```python
scaler = scaling('minmax')
```

```python
copied_features = pd.DataFrame(features, columns=features.columns)
```

```python
to_scale_features = copied_features[numeric_collums]
scaled_features = scaler.fit_transform(to_scale_features)
copied_features[numeric_collums] = scaled_features
copied_features
```

## Correlation Analysis (Task F)

```python
stddev = copied_features.std()
zero_stdev = stddev[stddev == 0].index
features_correlations = copied_features.drop(zero_stdev, axis=1)
```

```python
target_series = labels['Attrition']
correlations = features_correlations.corrwith(target_series)
correlations
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 15))
sns.barplot(x=correlations.values, y=correlations.index)
plt.title('Correlation of Each Feature with Target (Attrition)')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.show()
```

```python
correlations_matrix = pd.DataFrame(correlations, columns=['Correlation'])
plt.figure(figsize=(10, 15))
sns.heatmap(correlations_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation of Each Feature with Target (Attrition)')
plt.show()
```

```python
correlations = correlations.abs()
correlations = correlations.nlargest(20)
correlations
```

```python
train_features = copied_features[correlations.index]
```

## 1D Scatter Plot

```python
import numpy as np 

class_0 = features_correlations.loc[labels["Attrition"] == 0]
class_1 = features_correlations.loc[labels["Attrition"] == 1]

def plot_correlation(collum_name):
    plt.plot(class_0[collum_name], np.zeros_like(class_0[collum_name]), 'o', color='cyan', label='Class 0')
    plt.plot(class_1[collum_name], np.zeros_like(class_1[collum_name]), 'x', color='green', label='Class 1')
    plt.legend()
    plt.xlabel(collum_name)
    plt.title('1D Scatter Plot of sepal_width by Numeric Classes')
    plt.show()
```

```python
for collum in correlations.index:
    plot_correlation(collum)
```
